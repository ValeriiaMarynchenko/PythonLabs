## Написання та запуск коду. Налаштування PyCharm-Git-GitHub середовища 

Для ефективної, зокрема командної, роботи потрібно створений проект зв'язати з системою контроля версій (СКВ). Існують різні СКВ, найбільш популярною з яких наразі є [Git](https://git-scm.com/).
Для того, щоби налаштувати середовище PyCharm-Git-GitHub середовища виконуємо наступні дії:
1. Завантажуємо і налаштувуємо СКВ Git на своєму локальному комп'ютері.
1. Створюємо акаунт на ресурсі [GitHub](https://github.com/).
1. Створюємо репозиторій на GitHub у своєму акаунті.
1. Зв'язуємо його із створеним у PyCharm проєктом на python.

### Робота з віддаленим репозиторієм

Будь-які зміни в проєкті можуть бути додані в індекс Git-репозиторію і закомічені у репозиторій на локальній машині. Після того копія локального репозиторію може бути записана у віддалений репозиторій на GitHub.
Перелік необхідних команд Git, якшо викоростовуватидля цього консоль:

`git status` - перевірка статусу файлів

`git add .` - додавання всіх новостворених файлів у індекс

`git commit -m "коментар до коміту"` - додавання змін у локальний Git-репозиторій

`git push -u origin master` - відсилання поточної локальної копії Git-репозиторію на GitHub-репозиторій

Також все це можна зробити через меню PyCharm.

Маємо налаштоване середовище для ефективного написання та відлагодження коду на Python з можливістю збереження всіх версій проєкту як у локальному Git-середовищі, так і у віддаленому на GiHub.

### Контрольні запитання

* У чому відмінність інтерпретатора від компілятора? Python - компілятор чи інтерпретатор?

Компілятор перекладає програму на машинний код один раз, і відкомпільовану прогаму може відкрити будь-який користувач без доступу до коду та необхідності мати компілятор на власному ПК. Інтерпритатор пиконує "переклад" щоразу при виконанні програми, користувачу обов'язково необхідно мати встановлений інтерпритатор для відкриття прогами, а також у користувачів є доступ до коду програми. При інтерпритації не обов'язково щорозу "перекладати" увесь написаний код, це можна робити по одній команді. Python є мовою, що інтерпритується.

* Що таке IDE PyCharm?

IDE PyCharm - це кроссплатформенна інтегрована середовище розробки мови Python, розроблена компанією JetBrains на основі IntelliJ IDEA. Надає користувачеві комплекс засобів для написання коду та візуальний відладчик.

* Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ svn та git?

Система контролю версій - це система, що записує зміни до файлу або набір файлів протягом часу і дозволяє повернутися пізніше до певної версії. GIT зберігає метадані змін, а SVN – цілі файли. Це економить місце та час

* Що відбувається при виконанні команди `git add .`? 

При виконанні команди `git add .` відбувається додавання всіх новостворених файлів у індекс

* Що відбувається при виконанні команди `git commit`? 

При виконанні команди `git commit` відбувається додавання змін у локальний Git-репозиторій

* Якою git-командою виконується розміщення локальної копії git-репозиторію проєкта на віддаленому, що знаходиться на GitHub?

Відсилання поточної локальної копії Git-репозиторію на GitHub-репозиторій здійснється командою `git push -u origin master`
